```python
def sub_str(x):
    index = x.find(" C")
    return x[:index]

housing_raw["region_name"].apply(sub_str).head(10)   
```

```python
housing_raw["region_name"].apply(lambda x: x[: x.find(" C")]).head(10)
```

```python
housing_raw["region_name"].str.extract(r"(\w+)")
```

```python
ca_housing = (
    housing_raw
    .loc[housing_raw["region_name"].str.contains(r"\bCA\b", regex=True)]
    .assign(region_name = housing_raw["region_name"].apply(lambda x: x[0: x.find(" C")] ))
    .query("region_name in ['San Francisco', 'Los Angeles', 'San Diego']")
           )
```

```python
ca_housing
```

```python
ca_housing_markets = (
    ca_housing
    .set_index("period_begin")
    .loc[:,["region_name", "median_active_list_price"]]
    .sort_index()
    .pivot_table(
        index="period_begin",
        columns="region_name",
        values= "median_active_list_price"
    )
    .resample("Q")
    .mean()

)
```

```python
ca_housing_markets
```

```python
fig, ax = plt.subplots(figsize=(10,6))

ax.plot(
    ca_housing_markets.index, 
    ca_housing_markets["San Francisco"]/1000000,
    color="green",
    linewidth=2, 
    label= "San Francisco"
)

ax.plot(
    ca_housing_markets.index,
    ca_housing_markets["Los Angeles"]/1000000,
    color="purple",
    linewidth=2,
    label="Los Angeles"
)

fig.text(.15, .9,
         "San Francisco Prices Hit Hard By Covid...",
         fontsize=14

)
ax.text(18400, .6,
         "...and LA prices keep rising",
        fontsize=12

       )

ax.axvline(18341, color="black", linestyle="--")
        
         
ax.set_xlabel("Year")
ax.set_ylabel("Median Home Price (Millions)")         
ax.set_ylim(0, 1.6)
# ax.set_xlim(2017,2022)
ax.legend()
plt.show()
```
