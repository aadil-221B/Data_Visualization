# Matplotlib Plotting Examples

## Basic Plots:

1. **Line Plot**:
   - *Description*: Plotting a line connecting data points.
   - *Explanation*: Line plots are used to visualize data trends over continuous variables.
   - *Example*:
     ```python
     import matplotlib.pyplot as plt

     x = [1, 2, 3, 4, 5]
     y = [2, 3, 5, 7, 11]

     plt.plot(x, y)
     plt.xlabel('X-axis')
     plt.ylabel('Y-axis')
     plt.title('Line Plot Example')
     plt.show()
     ```

2. **Scatter Plot**:
   - *Description*: Plotting individual data points without connecting lines.
   - *Explanation*: Scatter plots are useful for visualizing relationships between two continuous variables.
   - *Example*:
     ```python
     import matplotlib.pyplot as plt

     x = [1, 2, 3, 4, 5]
     y = [2, 3, 5, 7, 11]

     plt.scatter(x, y)
     plt.xlabel('X-axis')
     plt.ylabel('Y-axis')
     plt.title('Scatter Plot Example')
     plt.show()
     ```

3. **Bar Plot**:
   - *Description*: Plotting bars to represent categorical data.
   - *Explanation*: Bar plots are commonly used to compare values across different categories.
   - *Example*:
     ```python
     import matplotlib.pyplot as plt

     categories = ['A', 'B', 'C', 'D']
     values = [25, 40, 30, 35]

     plt.bar(categories, values)
     plt.xlabel('Categories')
     plt.ylabel('Values')
     plt.title('Bar Plot Example')
     plt.show()
     ```

## Statistical Plots:

4. **Histogram**:
   - *Description*: Plotting the distribution of a continuous variable.
   - *Explanation*: Histograms show the frequency distribution of numerical data.
   - *Example*:
     ```python
     import numpy as np
     import matplotlib.pyplot as plt

     data = np.random.normal(loc=0, scale=1, size=1000)

     plt.hist(data, bins=30)
     plt.xlabel('Value')
     plt.ylabel('Frequency')
     plt.title('Histogram Example')
     plt.show()
     ```

5. **Box Plot**:

   - *Explanation*: Box plots display the distribution of data based on quartiles.

## Advanced Plots:

6. **Contour Plot**:
   - *Description*: Visualizing 3D surfaces in 2D.
   - *Explanation*: Contour plots represent 3D data on a 2D plane using contour lines.
   - *Example*:
     ```python
     import numpy as np
     import matplotlib.pyplot as plt

     x = np.linspace(-3, 3, 100)
     y = np.linspace(-3, 3, 100)
     X, Y = np.meshgrid(x, y)
     Z = np.sin(np.sqrt(X**2 + Y**2))

     plt.contour(X, Y, Z)
     plt.xlabel('X-axis')
     plt.ylabel('Y-axis')
     plt.title('Contour Plot Example')
     plt.show()
     ```

7. **Heatmap**:
   - *Description*: Visualizing data in a grid-like format.
   - *Explanation*: Heatmaps represent data values in a matrix with colors.
   - *Example*:
     ```python
     import numpy as np
     import matplotlib.pyplot as plt

     data = np.random.rand(10, 10)

     plt.imshow(data, cmap='viridis')
     plt.colorbar()
     plt.title('Heatmap Example')
     plt.show()
     ```

8. **3D Plot**:
   - *Description*: Visualizing data in 3D space.
   - *Explanation*: 3D plots represent data points in three-dimensional Cartesian space.
   - *Example*:
     ```python
     import numpy as np
     import matplotlib.pyplot as plt
     from mpl_toolkits.mplot3d import Axes3D

     x = np.linspace(-5, 5, 100)
     y = np.linspace(-5, 5, 100)
     X, Y = np.meshgrid(x, y)
     Z = np.sin(np.sqrt(X**2 + Y**2))

     fig = plt.figure()
     ax = fig.add_subplot(111, projection='3d')
     ax.plot_surface(X, Y, Z)
     ax.set_xlabel('X-axis')
     ax.set_ylabel('Y-axis')
     ax.set_zlabel('Z-axis')
     plt.title('3D Plot Example')
     plt.show()
     ```

````python
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

plt.contour(X, Y, Z)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Contour Plot Example')
plt.show()
````

6. **3D Plot with Scatter Plot**:
   - *Description*: Visualizing 3D surface and scatter plot together.
   - *Explanation*: Combining 3D plot and scatter plot to represent 3D surface and individual data points.
   - *Example*:
     ```python
     import numpy as np
     import matplotlib.pyplot as plt
     from mpl_toolkits.mplot3d import Axes3D

     x = np.linspace(-5, 5, 100)
     y = np.linspace(-5, 5, 100)
     X, Y = np.meshgrid(x, y)
     Z = np.sin(np.sqrt(X**2 + Y**2))

     fig = plt.figure(figsize=(10, 6))
     ax = fig.add_subplot(111, projection='3d')
     ax.plot_surface(X, Y, Z, alpha=0.8)
     ax.scatter(X[::10, ::10], Y[::10, ::10], Z[::10, ::10], color='r', marker='o', label='Data Points')
     ax.set_xlabel('X-axis')
     ax.set_ylabel('Y-axis')
     ax.set_zlabel('Z-axis')
     ax.set_title('3D Plot with Scatter Plot')
     ax.legend()
     plt.show()

**Histogram with Line Plot**:
   - *Description*: Plotting a histogram along with a line plot.
   - *Explanation*: Combining histogram and line plot to visualize data distribution and trend.`

````python
from scipy.stats import norm

# Generate Random data points, mean 30 std_dev 2 --> randomly choosen
data = np.random.normal(loc=30, scale=2,size=1000)

# Calculate mean and std_dev
mean = np.mean(data)

std_dev = np.std(data)


# Generate x_values for Gaussian Curve
x_values = np.linspace(min(data), max(data), 100)

# Compute Gaussian curve using mean and standard deviation of the data
gaussian_curve = norm.pdf(x_values, loc=mean, scale=std_dev)

# Plot histogram of data

plt.hist(data, bins=30, density=True, alpha=0.7, label="Data")

# Plot Gaussian curve

plt.plot(x_values,gaussian_curve,label="Gaussian Fit")

plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Gaussian Fit Example')
plt.legend()
plt.show()

plt.show()
````
